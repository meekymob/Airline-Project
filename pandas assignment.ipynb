```python
# Find the number of cancelled flight for every airline per day
```

```python
airline_per_day = df.groupby(['DAY','AIRLINE']).agg({"CANCELLED":"sum"}).sort_values(by='DAY').rename(columns={"CANCELLED":"Cancelled_Airline"})
```

```python
airlines = airline_per_day.reset_index()
```

```python
# Question 2
```

```python
## Finding the number of cancelled and diverted airlines
```

```python
cancelled_diverted_flight = df.groupby(['WEEKDAY','AIRLINE']).agg({"CANCELLED":"sum","DIVERTED": "sum"}).reset_index()
```

```python
cancelled_diverted_flight
```

```python
#Converting to Dataframe
```

```python
cancelled_diverted_flight = pd.DataFrame(cancelled_diverted_flight)
```

```python
# Percentage of Cancelled
```

```python
cancelled_diverted_flight['Percentage_of_Cancelled']=round((cancelled_diverted_flight['CANCELLED']/cancelled_diverted_flight['CANCELLED'].sum())*100,2)
```

```python
# Percentage of Diverted
```

```python
cancelled_diverted_flight['Percentage_of_DIVERTED']=round((cancelled_diverted_flight['DIVERTED']/cancelled_diverted_flight['DIVERTED'].sum())*100,2)
```

```python
# The number and the percentage of cancelled and diverted airlines
```

```python
df['Airtime'] = df['AIR_TIME']
```

```python
origin_destination_flight = df.groupby(['ORG_AIR','DEST_AIR']).agg({"AIRLINE":"count","CANCELLED":"count","AIR_TIME":"mean","Airtime":"var"}).reset_index().rename(columns={"AIR_TIME":"Airtime_Average","Airtime":"Airtime_Variance"})
```

```python
flight_details = pd.DataFrame(origin_destination_flight)
```

```python
flight_details['Flight_Percentage'] = round(flight_details['AIRLINE']/flight_details['AIRLINE'].sum()*100,2)
```
